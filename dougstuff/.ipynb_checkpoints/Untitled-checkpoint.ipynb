```python
def check_n_vars (a_df, n):
    """Pick n of the variables in the data frame and summarize values"""
    cols = a_df.columns.tolist()
    selection = np.random.randint(0, a_df.shape[1], n)
    for idx in selection:
        col = cols[idx]
        print(col)
        print(a_df[col].value_counts())
        print(f"Count of pandas missing values: {sum(a_df[col].isna())}")
        
```

```python
check_n_vars(df, 100)
```
