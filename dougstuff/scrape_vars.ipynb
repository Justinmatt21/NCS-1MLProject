```python
def convert_tr_to_list(table_row):
    output = []
    for td in table_row:
        output.append(td.text)
    return output

def convert_trs_to_dict(table_rows):
    table_conversion = {}
    for i, row in enumerate(table_rows):
        table_conversion[i] = convert_tr_to_list(row)
        
    return table_conversion

def convert_trs_to_dataframe(table_rows):
    tr_dict = convert_trs_to_dict(table_rows)
    col_names = tr_dict[0]
    data = pd.DataFrame(tr_dict).values.T[1:]
    df = pd.DataFrame(data, columns=col_names)
    df = df[df.Label != 'Missing Values']
    df = df[df.Label != 'Total']
    return df[['Value', 'Label']]

def convert_trs_to_mapping(table_rows):
    df = convert_trs_to_dataframe(table_rows)
    mapdict = {}
    for row in df.itertuples():
        mapdict[row.Value] = row.Label
    
    return mapdict
```

```python
study_variables = soup.find("div", {"id": "variableGroups"}).find("div", {"id": "varpane"}).find_all("li")
```

```python
#for study_var in study_variables:
#    print(study_var.text.split()[0])
    
var_names = [study_var.text.split()[0] for study_var in study_variables]
var_names
```

```python
var_urls = []
for study_var in study_variables:
    anchors = study_var.find_all("a", href=True)
    for a in anchors:
        var_urls.append(a['href'])

var_urls
```

```python
from functools import reduce
var_text = [study_var.text.split()[1:-3] for study_var in study_variables]
var_text = [reduce(lambda x,y: x+" "+y, text,'').strip() for text in var_text]
```

```python
len(var_text)
```

```python
vars_df = pd.DataFrame({
    'Name': var_names,
    'Text': var_text,
    'URL': var_urls
})
vars_df.head()
```

```python
vars_df.to_csv("variables.csv")
```

```python
# Define database and collection
db = client.ncs_db
collection = db.variable_maps
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
import time

def scrape_variable_info():
    
    variable_dict = {}
    
    for row in vars_df.itertuples():
        print(row.Name)
        try:
            browser.visit(row.URL)
            time.sleep(5)
            
            html = browser.html
            soup = BeautifulSoup(html, 'html.parser')

            table_rows = soup.find_all("tr")
        
            if table_rows:
                if len(table_rows) > 0: #redundant
                    variable_dict[row.Name] = convert_trs_to_mapping(table_rows)
                    
                    # Dictionary to be inserted as a MongoDB document
                    post = {
                        'Name': row.Name,
                        'map': variable_dict[row.Name]
                    }
                    collection.insert_one(post)
                    
        except Exception as e:
            print(e)
            
    return variable_dict    
```

```python
var_dict = scrape_variable_info()
```

```python
import json
with open('variables.json', 'w') as fp:
    json.dump(var_dict, fp)
```
